Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
# generate data
x1 = range(0,100)
y1 = [math.sqrt(n) for n in x1]

# plot the data
plt.plot(x1,y1)

# customize the chart
plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)

plt.grid(True, ls = '--');
```

```python
# generate data
x2 = range(0,20)
y2 = [n**3 for n in x2]

#plot data
plt.plot(x2,y2)

#customize the chart
plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--');
```

```python
# generate data
x3 = [x * 0.1 for x in range(0,150)]
y3 = [2 ** n for n in x3]

# plot data
plt.plot(x3,y3)

#customize the chart
plt.title('$ y = 2^x$')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.grid(True, ls = '--')
# plt.xticks(range(0, 16))
;
```

```python
# specify range of x values
x4 = [x * 0.1 for x in range(0,150)]
y4 = [1/(n+1) for n in x4]


# plot the data
plt.plot(x4,y4)


# Customize the plot
# Add title
plt.title('$ y = 1/(x+1)$')

# Add x and y labels
plt.xlabel('$x$')
plt.ylabel('$1/(x+1)$')


plt.grid(True, ls = '--')

plt.show()
```

```python
plt.figure(figsize = (10,6))

#subplot 1
plt.subplot(2,2,1)

plt.plot(x1,y1)

plt.title('$ y = \sqrt{x}$')
plt.xlabel('$x$')
plt.ylabel('$\sqrt{x}$')
plt.xlim(0,100)
plt.ylim(0, 10)
plt.grid(True, ls = '--')

#subplot 2
plt.subplot(222)

plt.plot(x2,y2)

plt.title('$ y = x^3$')
plt.xlabel('$x$')
plt.ylabel('$x^3$')
plt.xticks(range(0,20,5))
plt.xlim(0,20)
plt.ylim(0)
plt.grid(True, ls = '--')


#subplot 3
plt.subplot(223)

plt.plot(x3,y3)
plt.grid(True, ls = '--')
plt.xlabel('$x$')
plt.ylabel('$2^x$')
plt.title('$y = 2^x$')

#subplot 4
plt.subplot(224)

plt.plot(x4,y4)
plt.title(r'$ y = \frac{1}{x+1}$')
plt.xlabel('$x$')
plt.ylabel(r'$\frac{1}{x+1}$')
plt.grid(True, ls = '--')
plt.xlim(0,15)
plt.ylim(0)


plt.tight_layout()
```
