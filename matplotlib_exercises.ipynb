Use matplotlib to plot the following equation:



```python
import matplotlib.pyplot as plt 
```

```python
import math
```

```python
x = [0,1,2,3,4,5,6,7]
y = [x ** 2 - x + 2 for x in x]
```

```python
plt.plot(x,y, color = 'm', marker = 'v')

plt.annotate('This is 0,0', xy = (0,0), xytext= (4,4), arrowprops = dict(facecolor = 'blue'))
```

Create and label 4 separate charts for the following equations (choose a range for x that makes sense):

```python
x = [1,2,3,4]
y1 = [math.sqrt(x) for x in x]

y2 = [x ** 3 for x in x]

y3 = [x ** x for x in x]

y4 = [1 /(x+1) for x in x]
```

```python
plt.figure(figsize = (8,8))

plt.subplot(221)
plt.plot(x,y1)

plt.subplot(222)
plt.plot(x,y2)

plt.subplot(223)
plt.plot(x,y3)

plt.subplot(224)
plt.plot(x,y4)
```

```python
plt.figure(figsize = (10,10))


plt.title('question 2 chart')

plt.plot(x,y1, color = 'grey', ls = '--', marker = '^', label = 'y1')


plt.plot(x,y2, color = 'blue', marker = 'o', label = 'y2')


plt.plot(x,y3, color = 'green', marker = '*', label = 'y3')


plt.plot(x,y4, color = 'orange', label = 'y4')


plt.legend(loc = 'upper center')
```
